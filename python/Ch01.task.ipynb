**유연성:** 동적 타입 언어는 변수의 타입을 미리 지정하지 않아도 되므로 코드 작성이 더 자유롭고 유연합니다.

**런타임 오류 발생 가능성:** 프로그램이 실행되는 도중에야 타입 오류를 발견할 수 있으므로 디버깅이 어렵습니다.

**성능 문제:** 컴파일러가 타입을 미리 알 수 없으므로 최적화가 제한적일 수 있습니다.

<span style="font-size:80%">[문제4]<br><br></span>
```프로세스와 스레드의 차이점을 설명해주세요```

<span style="color:red">답변</span>

독립성: 프로세스 는 완전히 독립적으로 실행되지만, 스레드 는 같은 프로세스 내에서 실행되므로 서로 영향을 줄 수 있습니다. 

자원 공유: 프로세스 간에는 자원을 공유하지 않지만, 스레드 는 메모리 같은 자원을 공유할 수 있습니다.


정답:

아래 4가지 특징을 이용해 차이를 알아보겠습니다.

1. **독립성:** 프로세스는 독립적인 실행 단위입니다. 각 프로세스는 별도의 주소 공간에서 실행되며, 한 프로세스의 작업이 다른 프로세스에 직접적인 영향을 미치지 않습니다. 반면에 스레드는 한 프로세스 내에서 실행되는 여러 실행 경로로, 같은 프로세스 내의 다른 스레드와 메모리와 자원을 공유합니다.
2. **자원 공유:** 프로세스는 독립적인 자원을 가지며, 다른 프로세스의 자원에 접근하려면 별도의 통신 방법(IPC, Inter-Process Communication)이 필요합니다. 반면에 스레드는 같은 프로세스 내에서 메모리 영역(힙 영역)과 자원을 공유하기 때문에, 스레드 간의 통신이 더 효율적입니다.
3. **생성 및 종료 비용:** 프로세스의 생성 및 종료 비용은 스레드에 비해 상대적으로 높습니다. 프로세스를 생성하거나 종료할 때 운영 체제는 독립적인 메모리 공간과 자원을 할당하거나 회수해야 하기 때문입니다. 반면에 스레드의 생성 및 종료 비용은 상대적으로 낮습니다.
4. **오류의 영향 범위:** 프로세스 내에서 오류가 발생하면 그 프로세스 내에서만 영향을 미치지만, 스레드에서 오류가 발생하면 같은 프로세스 내의 다른 스레드에도 영향을 미칠 수 있습니다.

이러한 차이점들로 인해 프로세스와 스레드는 각각 다른 상황과 요구 사항에 맞게 사용됩니다. 예를 들어, 자원을 공유해야 하는 경우나 효율적인 통신이 필요한 경우에는 스레드를, 독립적인 작업이 필요하거나 격리가 중요한 경우에는 프로세스를 사용하는 것이 적절할 수 있습니다.

<span style="font-size:80%">[문제5]<br><br></span>
```Python 2버전대와 3버전대의 차이를 성명해주세요```

<span style="color:red">답변</span>

두 버전은 print 명령어 문법이 다릅니다.

그저 문법적인 차이라서, 대수롭지 않게 여길 수 있기에 파이썬의 기능에 영향을 미치지 않습니다.
그렇지만 print 'A'라고 하는 것과 print('A')는 반드시 알고 넘어가야 할 차이입니다.

기본적으로 Python 3에서는 print 명령어가 print() 기능으로 대체되었습니다.

예를 들어, Python 2에서는 print "hello" 지만 Python 3에서는 print("hello")입니다

Python 3가 향상된 정수 나눗셈 기능을 가지고 있습니다.

Python 2에서는 정수에 나눗셈을 하면 가장 가까운 정수까지 반올림을 합니다. 소수점이 있으면 반올림하지 않습니다.

하지만 Python 3에서는 소수점 뒤에 0을 붙일 걱정할 필요없게 반올림 해주지 않습니다.

Python 3가 훨씬 직관적이고 처음 Python 프로그래밍을 배우는 사람에게 쉽습니다.



정답:

파이썬 2와 파이썬 3는 같은 프로그래밍 언어지만, 여러 가지 중요한 차이점이 있습니다. 파이썬 3는 파이썬 2의 몇몇 기능을 개선하고, 더 효율적이고 간결한 언어 디자인을 목표로 개발되었습니다. 주요 차이점은 다음과 같습니다:

1. **문법 차이:** 파이썬 2와 3는 몇몇 문법적인 차이가 있습니다. 가장 대표적인 예로는 print 함수를 들 수 있습니다. 파이썬 2에서는 'print "Hello, World!"'와 같이 괄호 없이 사용하지만, 파이썬 3에서는 'print("Hello, World!")'와 같이 괄호를 사용합니다.
2. **문자열 처리:** 파이썬 2에서는 일반 문자열이 ASCII 기반으로 처리되며, 유니코드 문자열은 별도로 'u' 접두사를 붙여서 'u"문자열"' 형태로 처리합니다. 반면 파이썬 3에서는 모든 문자열이 유니코드로 처리되며, 바이트 문자열은 별도로 'b' 접두사를 붙여서 'b"문자열"' 형태로 처리합니다.
3. **정수 나눗셈:** 파이썬 2에서는 정수 나눗셈의 결과도 정수로 나옵니다. 예를 들어, '3 / 2'는 '1'이 됩니다. 반면 파이썬 3에서는 정수 나눗셈의 결과가 실수로 나옵니다. 즉, '3 / 2'는 '1.5'가 됩니다.
4. **지원 기간:** 파이썬 2의 공식 지원은 2020년에 종료되었습니다. 따라서 보안 업데이트나 버그 수정 등이 더 이상 제공되지 않습니다. 반면 파이썬 3는 계속해서 업데이트되고 있으며, 새로운 기능들이 추가되고 있습니다.

이러한 차이점들로 인해, 현재는 파이썬 3를 사용하는 것이 추천됩니다. 기존에 파이썬 2로 작성된 코드를 파이썬 3로 이전하는 작업도 필요하게 되었습니다.

<span style="font-size:80%">[문제6]<br><br></span>
```Python에서 메모리 관리는 어떻게 이루어지는지 설명해주세요```

<span style="color:red">답변</span>

먼저, 파이썬은 자동으로 메모리를 관리하는 고급 프로그래밍 언어입니다. 이것은 개발자가 직접 메모리 할당 및 해제를 관리하지 않고도 메모리 관리를 자동으로 처리할 수 있다는 것을 의미합니다.

파이썬에서 메모리 관리는 참조 계수
referencecounting기반으로 이루어집니다. 이는 객체가 생성될 때마다 그 객체를 참조하는 변수의 개수를 추적하는 방식입니다. 객체를 참조하는 변수의 개수가 0이되면, 해당 객체의 메모리는 자동으로 해제됩니다.


정답:

파이썬에서의 메모리 관리는 주로 가비지 컬렉션(Garbage Collection)과 참조 카운팅(Reference Counting)이라는 두 가지 메커니즘을 통해 이루어집니다.

1. **참조 카운팅(Reference Counting):** 파이썬의 모든 객체는 참조 카운트를 가지고 있습니다. 이는 해당 객체를 참조하는 다른 객체의 수를 나타냅니다. 예를 들어, 어떤 객체를 변수에 할당하거나 리스트에 추가하면 해당 객체의 참조 카운트가 증가합니다. 반대로 객체의 참조를 제거하면 참조 카운트가 감소합니다. 참조 카운트가 0이 되면, 즉 객체를 참조하는 것이 없게 되면, 그 객체는 메모리에서 해제됩니다.
2. **가비지 컬렉션(Garbage Collection):** 파이썬의 가비지 컬렉션은 주로 순환 참조(Circular Reference)를 처리하기 위해 사용됩니다. 순환 참조란 서로가 서로를 참조하는 상황을 말하며, 이런 경우에는 참조 카운트가 0이 되지 않아 메모리에서 해제되지 않는 문제가 발생할 수 있습니다. 파이썬의 가비지 컬렉터는 이런 순환 참조를 탐지하고 처리하여, 더 이상 사용되지 않는 객체를 메모리에서 해제합니다.

이 외에도 파이썬은 메모리 풀(Memory Pool)을 사용하여 작은 메모리 블록의 할당과 해제를 효율적으로 처리합니다. 이는 메모리의 단편화를 줄이고 성능을 향상시키는 데 도움이 됩니다.

이렇게 파이썬은 다양한 메커니즘을 통해 메모리를 효과적으로 관리하며, 이로 인해 개발자는 메모리 관리에 대해 신경 쓰지 않고도 편리하게 코드를 작성할 수 있습니다.
